2-Load and Preprocess the Data

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load the dataset
df = pd.read_csv('color_data.csv')

# Extract features and labels
X = df[['R', 'G', 'B']]
y = df['Color']

# Normalize the RGB values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

3-Train and Evaluate Models

Draw histograms of the target in train and test sets

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Draw histograms of the target in train and test sets
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

# Plot for training set
sns.countplot(x=y_train, ax=ax[0])
ax[0].set_title('Train Set')
ax[0].set_xlabel('Color')
ax[0].set_ylabel('Count')
ax[0].tick_params(axis='x', rotation=45)

# Plot for test set
sns.countplot(x=y_test, ax=ax[1])
ax[1].set_title('Test Set')
ax[1].set_xlabel('Color')
ax[1].set_ylabel('Count')
ax[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

*Random Forest*

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Initialize and train the Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Make predictions
rf_predictions = rf_model.predict(X_test)

# Evaluate the model
print("Random Forest Accuracy:", accuracy_score(y_test, rf_predictions))
print("Random Forest Classification Report:\n", classification_report(y_test, rf_predictions))
```

*K-Nearest Neighbors (KNN)*

```python
from sklearn.neighbors import KNeighborsClassifier

# Initialize and train the KNN model
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X_train, y_train)

# Make predictions
knn_predictions = knn_model.predict(X_test)

# Evaluate the model
print("KNN Accuracy:", accuracy_score(y_test, knn_predictions))
print("KNN Classification Report:\n", classification_report(y_test, knn_predictions))
```

4-Hyperparameter Tuning

*Random Forest Hyperparameter Tuning*

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load your dataset
# Assuming 'color_data.csv' is uploaded and contains the data
df = pd.read_csv('color_data.csv')
X = df[['R', 'G', 'B']]
y = df['Color']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Normalize the RGB values
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Hyperparameter tuning for Random Forest
n_estimators = range(10, 201, 10)  # Number of trees in the forest
max_depths = [None, 10, 20, 30, 40, 50]  # Maximum depth of the trees
best_accuracy = 0
best_params = {}

accuracies = []

for n in n_estimators:
    for max_depth in max_depths:
        rf = RandomForestClassifier(n_estimators=n, max_depth=max_depth, random_state=42)
        rf.fit(X_train_scaled, y_train)
        y_pred = rf.predict(X_test_scaled)
        accuracy = accuracy_score(y_test, y_pred)
        accuracies.append((n, max_depth, accuracy))

        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_params = {'n_estimators': n, 'max_depth': max_depth}

# Convert results to DataFrame for easy plotting
results_df = pd.DataFrame(accuracies, columns=['n_estimators', 'max_depth', 'accuracy'])

# Plotting accuracies for different hyperparameters
fig, ax = plt.subplots(figsize=(12, 6))
for depth in max_depths:
    subset = results_df[results_df['max_depth'] == depth]
    ax.plot(subset['n_estimators'], subset['accuracy'], marker='o', label=f'Max Depth={depth}')

ax.set_title('Random Forest: Accuracy vs. Number of Estimators')
ax.set_xlabel('Number of Estimators')
ax.set_ylabel('Accuracy')
ax.legend(title='Max Depth')
ax.grid(True)
plt.show()

print("Best Parameters for Random Forest:")
print(best_params)
print("Best Accuracy:", best_accuracy)
```

*KNN Hyperparameter Tuning*

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Sample data
import pandas as pd

# Assuming you have uploaded 'color_data.csv' and loaded it
df = pd.read_csv('color_data.csv')
X = df[['R', 'G', 'B']]
y = df['Color']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Normalize the RGB values
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Number of training samples
n_train_samples = X_train_scaled.shape[0]

# Set range of k values, ensuring k <= number of training samples
k_values = range(1, min(100, n_train_samples) + 1)
accuracies = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    y_pred = knn.predict(X_test_scaled)
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

# Plotting accuracies for different K values
plt.figure(figsize=(12, 6))
plt.plot(k_values, accuracies, marker='o')
plt.title('K-Nearest Neighbors: Accuracy vs. K')
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Accuracy')
plt.xticks(range(1, min(100, n_train_samples) + 1, 5))
plt.grid(True)
plt.show()

# Draw histograms of the target in train and test sets
import seaborn as sns

fig, ax = plt.subplots(1, 2, figsize=(12, 4))

sns.countplot(x=y_train, ax=ax[0])
ax[0].set_title('Train Set')
ax[0].set_xlabel('Color')
ax[0].set_ylabel('Count')
ax[0].tick_params(axis='x', rotation=45)

sns.countplot(x=y_test, ax=ax[1])
ax[1].set_title('Test Set')
ax[1].set_xlabel('Color')
ax[1].set_ylabel('Count')
ax[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
